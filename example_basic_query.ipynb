# Basic Query Example

This notebook shows the minimal text-query flow through the pipeline. Update the config path and source file paths for your environment before running the cell.


```python
# Notebook example: basic query flow
from pathlib import Path
import sys

project_root = Path.cwd()
if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))

from src.config.loader import load_config
from src.pipeline import PipelineOrchestrator

config = load_config(project_root / 'config' / 'config.yaml')
orchestrator = PipelineOrchestrator.from_config(config)

result = orchestrator.answer(
    query='Summarize the key points from the provided documents.',
    source_paths=['./data/example.pdf', './data/notes.md'],
    ingest_sources=True,
)

print(result.response_text)
print('Success:', result.success)
print('Retrieved chunks:', len(result.retrieved_chunks))
```
