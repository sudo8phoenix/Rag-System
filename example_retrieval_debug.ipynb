# Retrieval Debug Example

Use this notebook to inspect the returned chunks, scores, and ingest warnings for a single query.


```python
# Notebook example: retrieval debug flow
from pathlib import Path
import sys

project_root = Path.cwd()
if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))

from src.config.loader import load_config
from src.pipeline import PipelineOrchestrator

config = load_config(project_root / 'config' / 'config.yaml')
orchestrator = PipelineOrchestrator.from_config(config)

result = orchestrator.answer(
    query='Which document mentions the sales forecast?',
    source_paths=['./data/sales_report.md', './data/sales.csv'],
    ingest_sources=True,
    top_k=5,
)

print('Success:', result.success)
print('Error stage:', result.error_stage)
print('Ingest errors:', result.ingest_errors)
print('Retrieved chunks:')
for item in result.retrieved_chunks:
    chunk = item.get('chunk', {})
    print('-' * 80)
    print('chunk_id:', chunk.get('chunk_id'))
    print('source:', chunk.get('source_doc', {}).get('filename'))
    print('score:', item.get('score'))
    print('metadata:', chunk.get('metadata', {}))
    print(chunk.get('text', '')[:500])
```
