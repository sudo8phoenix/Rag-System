# Batch Ingest Example

This notebook demonstrates how to ingest several files in one call and inspect the returned ingestion summary.


```python
# Notebook example: batch ingest flow
from pathlib import Path
import sys

project_root = Path.cwd()
if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))

from src.config.loader import load_config
from src.pipeline import PipelineOrchestrator

config = load_config(project_root / 'config' / 'config.yaml')
orchestrator = PipelineOrchestrator.from_config(config)

source_paths = [
    './data/reports/quarterly_report.pdf',
    './data/reports/sales_summary.md',
    './data/reports/metrics.csv',
]

ingestion = orchestrator.ingest_documents(source_paths)
print('Documents:', len(ingestion.documents))
print('Chunks:', len(ingestion.chunks))
print('Errors:', ingestion.errors)
```
