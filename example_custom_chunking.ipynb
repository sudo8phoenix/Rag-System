# Custom Chunking Example

This notebook shows how to configure chunking at the document-format level and inspect the resulting chunks.


```python
# Notebook example: custom chunking flow
from pathlib import Path
import sys

project_root = Path.cwd()
if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))

from src.chunkers.registry import ChunkingRegistry
from src.config.settings import AppConfig, ChunkingConfig, FormatChunkingConfig
from src.models.document import Document

config = AppConfig(
    chunking=ChunkingConfig(
        strategy='paragraph',
        chunk_size=250,
        chunk_overlap=30,
        per_format={
            'md': FormatChunkingConfig(strategy='heading_hierarchy'),
            'csv': FormatChunkingConfig(strategy='row_based', rows_per_chunk=10),
        },
    )
)

document = Document(
    text='''# Section One
Some introductory text.

## Section Two
More content that should stay grouped under the heading.
''',
    filename='sample.md',
    source_type='md',
    original_metadata={'format': 'md'},
)

registry = ChunkingRegistry(config)
chunks = registry.chunk_document(document)

for chunk in chunks:
    print(chunk.chunk_id, chunk.strategy_used)
    print(chunk.text)
    print(chunk.metadata)
    print('---')
```
